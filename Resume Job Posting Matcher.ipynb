```python
#Import the library
import docx2txt
```

```python
#Store the resume in a variable
resume = docx2txt.process('CM_resume.docx')

print(resume)
```

```python
#upload the job description
job_description = docx2txt.process('JobDescription.docx')

print(job_description)
```

```python
# Remove punctuation and common words 
import nltk
import string
from nltk.corpus import stopwords
```

```python
string.punctuation
```

```python
#Lets make a function that takes care of all this for our dataframe of spam and ham
def text_process(mess):
    """"
    1. remove punc
    2. remove stop words
    3. return list of clean text words
    """
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    return [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
```

```python
# a list of text
text=[resume,job_description]
```

```python
text_process(text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()
```

```python
print(cv)
```

```python
count_matrix = cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#PRINT the similarity socres
print('\nSimilarity Scores:')
print(cosine_similarity(count_matrix))
```

```python
#Get the match percentage
matchPercentage = cosine_similarity(count_matrix)[0][1]*100
matchPercentage = round(matchPercentage,2)
print('Match Percentage is: '+ str(matchPercentage))
```
