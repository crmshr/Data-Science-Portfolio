```python
import urllib.request
from bs4 import BeautifulSoup
import os 
```

```python
def make_soup(url):
    webpage = urllib.request.urlopen(url)
    soup_data = BeautifulSoup(webpage,'html.parser')
    return soup_data
```

```python
# Scrape basketball reference for the URL path chosen
playerdatasaved=""
soup = make_soup('https://www.basketball-reference.com/leagues/NBA_2020_standings.html')

# For loop for all table data.  tr = table rows

for read_data in soup.findAll('tr'):
    playerdata = ""
    for data in read_data.findAll('td'):# td = table data 
        playerdata = playerdata+','+data.text
    if len(playerdata) != 0:
        playerdatasaved = playerdatasaved + "\n" + playerdata[1:]
```
